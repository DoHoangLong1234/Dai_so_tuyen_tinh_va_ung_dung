```python
# BÀI TẬP 1
import numpy as np
A = np.array([[2, 1, 3],[0, 1, 4], [5, 2, 1]])
print("Ma trận A:")
print(A)
det = np.linalg.det(A)
if det != 0:
    print("Ma trận khả nghịch vì định thức khác 0:", det)
    A_inv = np.linalg.inv(A)
text = "Do Hoang Long"
def text_to_vector(text):
    ascii_vals = [ord(c) for c in text]
    while len(ascii_vals) % 3 != 0:
        ascii_vals.append(32) 
    return np.array(ascii_vals).reshape(-1, 3).T
def encrypt(text_vec, matrix):
    return np.dot(matrix, text_vec)
vec = text_to_vector(text)
encrypted = encrypt(vec, A)
print("Mã hóa:")
print(encrypted)
def decrypt(enc_vec, matrix_inv):
    return np.dot(matrix_inv, enc_vec)
decrypted = decrypt(encrypted, A_inv).round().astype(int)
decrypted_text = ''.join([chr(num) for num in decrypted.T.flatten()])
print("Giải mã:")
print(decrypted_text)
```

```python
# BÀI TẬP 2
from fractions import Fraction
def continued_fraction_to_fraction(coeffs):
    result = Fraction(coeffs[-1])
    for c in reversed(coeffs[:-1]):
        result = c + Fraction(1, result)
    return result.numerator, result.denominator
coeffs = [3, 1, 2]
pn, qn = continued_fraction_to_fraction(coeffs)
print("Kết quả pn/qn:", pn, "/", qn)
from sympy import Matrix
def verify_matrix_formula(coeffs):
    M = Matrix([[1, 0], [0, 1]])
    for c in coeffs:
        M = M * Matrix([[c, 1], [1, 0]])
    return M[0, 0], M[1, 0]
pn2, qn2 = verify_matrix_formula(coeffs)
print("Kiểm chứng bằng ma trận:", pn2, "/", qn2)
```
